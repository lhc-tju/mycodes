```python
import pandas as pd
import numpy as np

def get_dir_filenames(path,ftype):
    '''获取目录下所有指定格式的文件名'''
    import os
    names = []
    for root, dirs, files in os.walk(path):
        for file in files:
            if ftype in file:
                names.append(file)
    return names

fnames = get_dir_filenames('论文数据/知网统计学数据/csv','.csv')
for name in fnames:
    try:
        data = pd.read_csv('论文数据/知网统计学数据/csv/'+name,engine='python')
        all_data.append(data)
    except:
        print(name)

df = pd.concat(all_data)
print(len(df))
df = df.drop_duplicates()
print(len(df))
df.to_excel('论文数据/知网-统计学-汇总.xlsx')
```
