```python
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
import re

colleges = pd.read_excel('高校名单.xlsx',sheet_name='全部高校')
colleges = set(colleges['高校名称'])
```

```python
#index_url = 'http://fz.people.com.cn/skygb/sk/index.php/Index/seach?gzdw=首都经济贸易大学&p=1'
index_url = 'http://fz.people.com.cn/skygb/sk/index.php/Index/seach?gzdw='

alldata = []
for col in colleges:
    i = 0
    while True:
        try:
            url = index_url + col + '&p=' + str(i+1) # 如果没有此url,则表明退出
            res = requests.get(url)
            soup = BeautifulSoup(res.text,'html5lib')
            items = soup.select('.jc_a')[0].select('tbody')[0].select('tr')
            for item in items:
                save_data = []
                each_i = item.select('td')
                for j in each_i:
                    save_data.append(j.text)
                alldata.append(save_data)
            i+=1
        except:
            print(col + ' 采集完毕，共'+str(i)+'页')
            break
    
columns = ['项目批准号','项目类别','学科分类','项目名称','立项时间','项目负责人','专业职务','工作单位','单位类别','所在省区市','所属系统','成果名称','成果形式','成果等级','结项时间','结项证书号','出版社','出版时间','作者','获奖情况']
df = pd.DataFrame(alldata,columns=columns)
len(df)
```

```python
index_url = 'http://fz.people.com.cn/skygb/sk/index.php/Index/seach?gzdw='

alldata = []
for col in colleges:
    url = index_url + col #第一页
    res = requests.get(url)
    soup = BeautifulSoup(res.text,'html5lib')
    text = soup.select('.page')[0].text
    page_counts = int(re.findall('(\d+) 页',text)[0])
    for i in range(page_counts): #遍历所有页面
        url = index_url + col + '&p=' + str(i+1)
        res = requests.get(url)
        soup = BeautifulSoup(res.text,'html5lib')
        items = soup.select('.jc_a')[0].select('tbody')[0].select('tr')
        for item in items:
            save_data = []
            each_i = item.select('td')
            for j in each_i:
                save_data.append(j.text)
            alldata.append(save_data)
    print(col + ' 采集完毕')
     
```

```python
columns = ['项目批准号','项目类别','学科分类','项目名称','立项时间','项目负责人','专业职务','工作单位','单位类别','所在省区市','所属系统','成果名称','成果形式','成果等级','结项时间','结项证书号','出版社','出版时间','作者','获奖情况']
df = pd.DataFrame(alldata,columns=columns)
df = df.drop_duplicates()
df.to_excel('高校国家社科基金项目.xlsx')
```
