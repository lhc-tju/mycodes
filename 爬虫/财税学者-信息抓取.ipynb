```python
import requests 
from bs4 import BeautifulSoup

url = 'http://www.gzcc.edu.cn/dxwyjxb/channels/8813.html'
index_url = 'http://www.gzcc.edu.cn'
#抓取url
def geturl(url,index_url):
    print('url')
    urls = []
    for i in range(1,4):
        if i == 1:
            n_url = url
        else:
            n_url = url + '_' + str(i)
        res = requests.get(url)
        res.encoding='utf-8'
        soup = BeautifulSoup(res.text,'lxml')
        result = soup.select('.news_list')
        names = result[0].find_all('a')
        for name in names:
            u = index_url + name['href']
            urls.append(u)
        
    return urls

def getpage(urls):
    print('pages')
    all_messages = []
    for url in urls:
        print(url)
        page = requests.get(url)
        print('test')
        page.encoding = 'utf-8'
        print('aa')
        u_soup = BeautifulSoup(page.text,'lxml')
        print('bb')
        u_re = u_soup.select('.content')
        print('cc')
        detail = u_re[0].text
        
        all_messages.append(detail.strip())
    return all_messages
    
urls = geturl(url,index_url)
all_messages = getpage(urls)
with open('贵州商学院.txt','rb') as f:
    for item in all_messages:
        f.writelines(item)
        f.write('\n')

print(all_messages)



'''
res = requests.get(url)
res.encoding='utf-8'
soup = BeautifulSoup(res.text,'lxml')
result = soup.select('.news_list')
names = result[0].find_all('a')
for item in names:
    data = ''
    scholar = item.text
    u = index_url + item['href']
    page = requests.get(u)
    page.encoding = 'utf-8'
    u_soup = BeautifulSoup(page.text,'lxml')
    u_re = u_soup.select('.content')
    detail = u_re[0].text
    #data = data + scholar + ',' + detail.strip()
    all_messages.append(detail.strip())
print(all_messages)
''' 
    
    
```
